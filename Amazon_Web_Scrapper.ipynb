```python
# Import Libraries/Modules
"""
 Term is interchangeable but Python Documentation or resources
 may refer to them as Modules rather than Libraries which
 System or Application based programming languages
 will classify them as Liberaries. Such as: C, C++, Java, etc.
"""

from bs4 import BeautifulSoup
import requests
import time
import datetime
```

```python
# Connect to Website and pull in data
# Relevant with Amazon's HTML and CSS layout as of 03/17/2022

URL = 'https://www.amazon.com/Funny-Data-Systems-Business-Analyst/dp/B07FNW9FGJ/ref=sr_1_3?dchild=1&keywords=data%2Banalyst%2Btshirt&qid=1626655184&sr=8-3&customId=B0752XJYNL&th=1'

# Refer to this link: https://httpbin.org/get
# To determine what the User Agent for your Computer is!
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

page = requests.get(URL, headers=headers)

soup1 = BeautifulSoup(page.content, "html.parser")

soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

title = soup2.find(id='productTitle').getText()

price = soup2.find(class_='a-offscreen').getText()


print(title)
print(price)
```

```python
# Clean up the data a little bit
# We are cleaning up any white space at the start and
#  end of the string variables.

price = price.strip()[1:]
title = title.strip()

print(title)
print(price)
```

```python
# Create a Timestamp for your output to track when data was collected

import datetime

today = datetime.date.today()

print(today)
```

```python
data = {
    'Title': title, 
    'Price': price, 
    'Date': today}

print(data)
```

```python
# Create DataFrame that will be written as a csv file.
import pandas as pd

df = pd.DataFrame([data])
print(df)
```

```python
# Create CSV and write headers and data into the file
df.to_csv("AmazonWebScraperDataset.csv", header=True, index=False)
```

```python
# Combine all of the above code into one function
def check_price():
    URL = 'https://www.amazon.com/Funny-Data-Systems-Business-Analyst/dp/B07FNW9FGJ/ref=sr_1_3?dchild=1&keywords=data%2Banalyst%2Btshirt&qid=1626655184&sr=8-3&customId=B0752XJYNL&th=1'

    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

    page = requests.get(URL, headers=headers)

    soup1 = BeautifulSoup(page.content, "html.parser")

    soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

    title = soup2.find(id='productTitle').getText()

    price = soup2.find(class_='a-offscreen').getText()

    price = price.strip()[1:]
    title = title.strip()

    import datetime

    today = datetime.date.today()
    
    data = [title, price, today]

    import csv 
    with open('AmazonWebScraperDataset.csv', 'a+', newline='', encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data)
```

```python
# Runs check_price after a set time and inputs data into your CSV
# Change from True to "len('AmazonWebScraperDataset.csv')  < some_number_of_lines - 1" or vice versa as needed.
# Staying with "while(True)" will continue the program until you STOP or terminate Notebook execution.
while(len('AmazonWebScraperDataset.csv') < 3): 
    check_price()
    time.sleep(15) # To stop code from running press the STOP button modify the "while()" line.
```
